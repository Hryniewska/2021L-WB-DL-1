```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Regularization Results

```python
runs = [
    {'model': 'model-f4fd5d70-4677-4c48-b143-9900624197c6', 'l1': 0, 'l2': 1e-4 },
    {'model': 'model-df6003fb-591e-4e4a-871c-b05cec08c4cf', 'l1': 0, 'l2': 1e-2 },
    {'model': 'model-7d7949e7-b6a5-4045-9bf0-925221e37f1c', 'l1': 1e-4, 'l2': 0 },
    {'model': 'model-a665552b-f99f-4a37-84ea-8ab704e231f3', 'l1': 1e-2, 'l2': 0}
]
```

```python
for run in runs:
    run['log'] = pd.read_csv('LungNet/' + run.get('model') + '/training.log')
```

```python
plt.rcParams["figure.figsize"] = (14,6)
for run in runs:
    log = run.get('log')
    label = 'L1: %s L2: %s' % (run.get('l1'), run.get('l2'))
    plt.plot(log['epoch'], log['val_waccOA'], label=label)
plt.legend()
plt.ylabel('Weighted accuracy')
plt.xlabel('Epoch')
plt.show()
```

Regularyzacja służy do zmniejszenia zjawiska przetrenowania. Ten model oznacza wszystko jedną klasą, dlatego regularyzacja mu nie pomoże. Bez znaczenia czy to L1 czy L2.
